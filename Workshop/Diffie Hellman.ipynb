## Diffie-Hellman key Exchange
Diffie-Hellman key exchange raises numbers to a selected power to produce decryption keys. The components of the keys are never directly transmitted, making the task of a would-be code breaker mathematically overwhelming
1
. The method doesn't share information during the key exchange. The two parties have no prior knowledge of each other, but the two parties create a key together


## Modulo Arithmetic

The easiest way to understand modular arithmetic is to think of it as finding the remainder of a number upon division by another number. 

For example, since both 15 and 27 leave the same remainder 3 when divided by 12

So 15 mod 12 = 3
And 27 mod 12 also = 3

You must first agree on 2 Numbers

Prime Modulus = 17  Generator = 3

```python
def calc_power(N, p):
    if p == 0:
        return 1
    return N * calc_power(N, p-1)

# Alices secret is not to be revealed for any reason
alice_secret = 5
generator = 3
a=(calc_power(generator, alice_secret))
```

```python

#Prime Modulus = b
b = 17
 
pub_alice = a % b
print(a, "mod", b, "=",
      pub_alice, sep=" ")
print("This number is now sent Publically to Bob for his next calculation",pub_alice)
```
